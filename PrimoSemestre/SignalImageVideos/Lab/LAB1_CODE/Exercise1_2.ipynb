```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Create a pandas DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 45],
    'Salary': [50000, 60000, 70000, 80000, 90000]
}
```

Create Database

```python
df1 = pd.DataFrame(data)
```

display it

```python
print(df1)
```


Accessing and modifying specific values in the DataFrame

Change Alice's age to 26

```python
df1.at[0, 'Age'] = 26
```

```python
print(df1)
```

Increase Bob's salary by 5000


```python
df1.at[1, 'Salary'] = 65000
print(df1.iloc[1])
```

Given data2 concatenate it with data1 and display the new results

```python
data2 = {
    'Name': ['Frank', 'Grace', 'Hannah', 'Ivy', 'Jack'],
    'Age': [30, 35, 40, 45, 50],
    'Salary': [75000, 85000, 95000, 105000, 115000]}

df2 = pd.DataFrame(data2)
```

```python
df = pd.concat([df1, df2], ignore_index=True)
print(df)
```

Sort the results by Salary, Age and Name

```python
print(df.sort_values(by=['Salary']))
```

```python
print(df.sort_values(by=['Age']))
```

```python
print(df.sort_values(by=['Name']))
```

```python
df_sorted = df.sort_values(by=['Salary', 'Age', 'Name'])
print(df_sorted)
```

Plot Age vs. Salary

```python
import matplotlib.pyplot as plt

df.plot(x='Age', y='Salary', kind='scatter')
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```
