## Objective: 

The objective of this exercise is to help beginners get familiar with NumPy, scipy, scikit-learn, and OpenCV while performing basic image processing and analysis tasks.

```python
#read image

import cv2

# Load the image
image = cv2.imread('imgs_lab/sample_image.png')
```

Display image

```python
import matplotlib.pyplot as plt

# Display the image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Sample Image')
plt.axis('off')  # Hide the axis
plt.show()
```

Convert to grayscale and display it

```python
# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')  # Hide the axis
plt.show()
```

Display every color channel [r,g,b]

```python
# Split the image into its color channels
blue_channel, green_channel, red_channel = cv2.split(image)

# Display the Blue channel
plt.figure(figsize=(10, 3))
plt.subplot(1, 3, 1)
plt.imshow(blue_channel, cmap='Blues')
plt.title('Blue Channel')
plt.axis('off')

# Display the Green channel
plt.subplot(1, 3, 2)
plt.imshow(green_channel, cmap='Greens')
plt.title('Green Channel')
plt.axis('off')

# Display the Red channel
plt.subplot(1, 3, 3)
plt.imshow(red_channel, cmap='Reds')
plt.title('Red Channel')
plt.axis('off')

plt.show()
```

Flip the image

```python
# Flip the image vertically
flipped_image = cv2.flip(image, 0)

# Display the flipped image
plt.imshow(cv2.cvtColor(flipped_image, cv2.COLOR_BGR2RGB))
plt.title('Flipped Image')
plt.axis('off')  # Hide the axis
plt.show()
```

Crop the image

```python
# Define the cropping coordinates (startY, endY, startX, endX)
startY, endY, startX, endX = 50, 200, 100, 300

# Crop the image using the defined coordinates
cropped_image = image[startY:endY, startX:endX]

# Display the cropped image
plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
plt.title('Cropped Image')
plt.axis('off')  # Hide the axis
plt.show()
```

Change color of a box in the image

```python
# Define the coordinates and size of the box (top-left corner and width/height)
box_startX, box_startY = 120, 80
box_width, box_height = 100, 100

# Change the color of the box to red (BGR format)
image[box_startY:box_startY+box_height, box_startX:box_startX+box_width] = [50, 100, 200]

# Display the modified image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Image with Red Box')
plt.axis('off')  # Hide the axis
plt.show()
```

## The Extra-Mile! 

Using the image rocky.jpg generate rocky_chess_pattern.png

```python
import numpy as np

image_path = 'imgs_lab/rocky.jpg'  
patch_size = 128  
# Load the image
rocky_image = cv2.imread(image_path)

# Get the dimensions of the image
height, width, _ = rocky_image.shape

# Create an empty canvas for the chess pattern
chess_pattern = np.zeros((height, width, 3), dtype=np.uint8)

# Iterate over the image and fill the chess pattern
for y in range(0, height, patch_size):
    for x in range(0, width, patch_size):
        # Determine the color of the patch
        if (y // patch_size) % 2 == (x // patch_size) % 2:
            chess_pattern[y:y+patch_size, x:x+patch_size] = rocky_image[y:y+patch_size, x:x+patch_size]

# Save the generated chess pattern image
cv2.imwrite('imgs_lab/rocky_chess_pattern.png', chess_pattern)

# Display the generated chess pattern image
plt.imshow(cv2.cvtColor(chess_pattern, cv2.COLOR_BGR2RGB))
plt.title('Rocky Chess Pattern')
plt.axis('off')  # Hide the axis
plt.show()
```
