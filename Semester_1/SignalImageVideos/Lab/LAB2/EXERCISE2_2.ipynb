## Objective: 

Learn how to apply various filters to an image using OpenCV, NumPy, and Matplotlib.

```python
#read image

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('imgs_lab/lena_color.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

Implement functions to apply various filters (gaussian blur, median blur and sobel edge detection) to the image

- https://docs.opencv.org/4.x/d4/d13/tutorial_py_filtering.html

- https://www.geeksforgeeks.org/python-program-to-detect-the-edges-of-an-image-using-opencv-sobel-edge-detection/


```python
def apply_gaussian_blur(image, kernel_size=(5, 5), sigma=0):
    return cv2.GaussianBlur(image, kernel_size, sigma)

def apply_median_blur(image, kernel_size=5):
    return cv2.medianBlur(image, kernel_size)

def apply_sobel_edge_detection(image, dx=1, dy=1, ksize=5):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    sobel = cv2.Sobel(gray, cv2.CV_64F, dx, dy, ksize=ksize)
    return cv2.convertScaleAbs(sobel)

# Apply the filters
gaussian_blur_image = apply_gaussian_blur(image)
median_blur_image = apply_median_blur(image)
sobel_edge_image = apply_sobel_edge_detection(image)

# Display the results
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title('Gaussian Blur')
plt.imshow(cv2.cvtColor(gaussian_blur_image, cv2.COLOR_BGR2RGB))

plt.subplot(1, 3, 2)
plt.title('Median Blur')
plt.imshow(cv2.cvtColor(median_blur_image, cv2.COLOR_BGR2RGB))

plt.subplot(1, 3, 3)
plt.title('Sobel Edge Detection')
plt.imshow(sobel_edge_image, cmap='gray')

plt.show()
```
